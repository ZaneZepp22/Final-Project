The data shows a three way tie for first place, but we wanted a true winner.

```python
nbh = pd.read_csv("Neighborhoods_.csv", index_col = "hood")
nbh.drop(['objectid', 'fid_blockg', 'statefp10', 'countyfp10', 'tractce10', 'geoid10', 'mtfcc10', 'funcstat10'],axis=1)
print("Mt. Oliver: ", nbh.loc["Mt. Oliver", "sqmiles"])
print("Ridgemont: ", nbh.loc["Ridgemont", "sqmiles"])
print("Regent Square: ", nbh.loc["Regent Square", "sqmiles"])
```

### Conclusion
